```python
import json
import pprint
import pickle
import pandas as pd
import numpy as np

class colors:
    '''Colors class:reset all colors with colors.reset; two
    sub classes fg for foreground
    and bg for background; use as colors.subclass.colorname.
    i.e. colors.fg.red or colors.bg.greenalso, the generic bold, disable,
    underline, reverse, strike through,
    and invisible work with the main class i.e. colors.bold'''
    reset='\033[0m'
    bold='\033[01m'
    disable='\033[02m'
    underline='\033[04m'
    reverse='\033[07m'
    strikethrough='\033[09m'
    invisible='\033[08m'
    class fg:
        black='\033[30m'
        red='\033[31m'
        green='\033[32m'
        orange='\033[33m'
        blue='\033[34m'
        purple='\033[35m'
        cyan='\033[36m'
        lightgrey='\033[37m'
        darkgrey='\033[90m'
        lightred='\033[91m'
        lightgreen='\033[92m'
        yellow='\033[93m'
        lightblue='\033[94m'
        pink='\033[95m'
        lightcyan='\033[96m'
    class bg:
        black='\033[40m'
        red='\033[41m'
        green='\033[42m'
        orange='\033[43m'
        blue='\033[44m'
        purple='\033[45m'
        cyan='\033[46m'
        lightgrey='\033[47m'

print(colors.bg.green, "SKk", colors.fg.red, "Amartya")
print(colors.bg.lightgrey, "SKk", colors.fg.red, "Amartya")
print(colors.reset)

```

```python
# Utils

def filter_debugger_data(file):
    filtered = []
    for event in file:
        x = json.loads(event["message"])["message"]
        if "Network.webSocketFrameReceived" in x["method"] or "Network.webSocketFrameSent" in x["method"]:
            filtered.append({"method": x["method"], "payloadData": json.loads(x["params"]["response"]["payloadData"])})
    return filtered

def search_by_name(data, keys):
    l = [element for element in data if element["payloadData"]["name"] in keys]
    return l

def search_complement_by_name(data, keys):
    l = [element for element in data if element["payloadData"]["name"] not in keys]
    return l

def search_by_request_id(data, request_id):
    l = []
    for element in data:
        try: 
            if element["payloadData"]["request_id"] == request_id:
                l.append(element)
        except:
            pass
    return l

def print_with_colors(data):
    for element in data:
        if "FrameReceived" in element["method"]:
            print(colors.bold, colors.fg.red, json.dumps(element, indent=4), colors.reset)
        elif "FrameSent" in element["method"]:
            print(colors.bold, colors.fg.green, json.dumps(element, indent=4), colors.reset)
```

# Handling candles

```python
# Get request_ids 
def get_request_ids(data):
    d = {}
    for element in data:
        if element["payloadData"]["name"] == "sendMessage" and element["payloadData"]["msg"]["name"] == "get-candles":
            d[element["payloadData"]["request_id"]] = {
                "active_id": element["payloadData"]["msg"]["body"]["active_id"],
                "size": element["payloadData"]["msg"]["body"]["size"],
            }
    return d
request_ids = get_request_ids(data)
request_ids
```

```python
# Get the data
def get_candles(data):
    candles_dict = {}
    for element in data:
        if element["payloadData"]["name"] == "candles":
            candles_dict[element["payloadData"]["request_id"]] = pd.DataFrame(element["payloadData"]["msg"]["candles"]).set_index("id")
    return candles_dict

candles = get_candles(data)   
candles
```

```python
binary_op_ids = { 7: 'AUDCAD', 943: 'AUDCHF', 101: 'AUDJPY', 944: 'AUDNZD', 99: 'AUDUSD', 107: 'CADCHF',
                106: 'CHFJPY', 108: 'EURAUD', 2: 'EURGBP', 4: 'EURJPY', 212: 'EURNZD', 1: 'EURUSD',
                104: 'GBPAUD', 102: 'GBPCAD', 103: 'GBPCHF', 3: 'GBPJPY', 947: 'GBPNZD', 5: 'GBPUSD', 8: 'NZDUSD',
                100: 'USDCAD', 72: 'USDCHF', 6: 'USDJPY', 168: 'USDNOK'}

assets = {}

# fill the assets dict
def fill_assets_dict(assets, request_ids, candles):
    for request_id in request_ids:
        active_id = request_ids[request_id]['active_id']
        size = request_ids[request_id]["size"]

        if active_id not in assets.keys():
            try:
                assets[active_id] = {
                    'name': binary_op_ids[active_id],
                    'data': {}
                }
            except:
                assets[active_id] = {
                    'name': '',
                    'data': {}
                }

        index = pd.MultiIndex.from_product([[size], candles[request_id].index.to_numpy()], names=['size', 'id'])
#             candles[request_id] = candles[request_id].set_index(index)
        candle = candles[request_id].set_index(index)

        if isinstance(assets[active_id]["data"], pd.DataFrame):
            assets[active_id]["data"] = assets[active_id]["data"].merge(candle, how='outer', on=candle.columns.values.tolist(), left_index=True, right_index=True)
#               This seems to keep duplicate values
#                 assets[active_id]["data"] = assets[active_id]["data"].append(candle) 
        else:
#                 assets[active_id]["data"] = candles[request_id]
            assets[active_id]["data"] = candle

        
            
fill_assets_dict(assets, request_ids, candles)

for key in assets:
    assets[key]['data'] = assets[key]['data'].sort_index()
```
